#### Grouping and Aggregating

```python
gr.groups
```

```python
gr.get_group('BE')
```

```python
gr.get_group('TE')
```

```python
gr.agg(sum)
```

```python
gr.agg(max)
```

```python
gr.mean(numeric_only=True)
```

#### Boolean Filtering

```python
df['marks'] > 60
```

```python
df[df['marks'] > 60]
```

```python
df[(df['marks'] > 60) & (df['class'] == 'BE')]
```

```python
x=df[(df['marks'] > 60) & (df['class'] == 'BE')]
```

#### Export the data
